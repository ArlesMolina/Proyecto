```python
def the_first():
    # Leer el archivo CSV
    df = pd.read_csv('netflix_titles.csv')

    # Retornar el primer renglón del DataFrame
    return df.head(0)

# Llamar la función
print(the_first())
```

```python
def row_number():
    # Leer el archivo CSV
    df = pd.read_csv('netflix_titles.csv')

    # Retornar el número de registros (filas) del DataFrame
    return df.shape[0]

# Llamar la función
print(row_number())
```

```python
def released_year(year):
    # Leer el archivo CSV
    df = pd.read_csv('netflix_titles.csv')

    # Filtrar las filas donde 'release_year' es igual al año dado
    movies = df[df['release_year'] == year]

    # Retornar las películas filtradas
    return movies

# Año de consulta
year_query = 2020

# Llamar la función
print(released_year(year_query))
```

```python
def origin_country(country):
    # Leer el archivo CSV
    df = pd.read_csv('netflix_titles.csv')

    # Filtrar las filas donde 'country' incluye el país dado
    movies = df[(df['country'].str.contains(country, na=False)) & (df['type'] == 'Movie')]

    # Retornar las películas filtradas
    return movies

# País de consulta
country_query = 'Mexico'

# Llamar la función
print(origin_country(country_query))
```

```python
# Leer el archivo CSV
df = pd.read_csv('netflix_titles.csv')

# Crear un nuevo DataFrame con las columnas deseadas
new_df = df[['type', 'title', 'country', 'release_year']]

# Mostrar el nuevo DataFrame
print(new_df)
```

```python
# Leer el archivo CSV
df = pd.read_csv('netflix_titles.csv')

# Agrupar por 'country' y 'type', y contar el número de producciones
productions_by_country = df.groupby(['country', 'type']).size().reset_index(name='count')

# Mostrar el resultado
print(productions_by_country)
```
